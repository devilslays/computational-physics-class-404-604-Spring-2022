```python
import numpy as np
from math import sin, pi
import matplotlib.pyplot as plt
from pylab import *
```

```python
constant = pi/2
n = 11
x=np.arange(-pi,pi,0.1)
y = constant + 2 * (sin(n*x)/n)
plt.plot(x,y)
plt.xlim(-pi,pi)
plt.ylim(-5,5)
plt.show()
```
