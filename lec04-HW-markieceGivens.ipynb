## Write a program to numerically evaluate the derivative of f(x) = cos x at an arbitrary position x

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Evaluate the derivative at position x using the forward difference scheme. Plot log10(Error) = log10[f'(numerically)-f'(analytic)] against log10(dx) for dx = 1e-16, 1e-15,..., 0.1 at x = $\pi$/4. You should find Error~$dx^n$; what is n? what is the optimal value for dx

#### Graphing log10(error) vs log10(dx)

```python
def derifw(y,x,dx):
    return(y(x+dx)-y(x))/dx

x = np.pi/4
dxarr = []
yarr = []
for dx in 10**np.arange(-16, -1, 0.1):
    y = np.abs(derifw(np.cos,x,dx)+np.sin(x)) ## this line of code is (Error) =[f'(numerically)-f'(analytic)]
    dxarr.append(dx) # this updates the values of dx into the dx array (taken from lecture)
    yarr.append(y) # this updates the yarr with y (the error) with the calculation (Error) = [f'(numerically)-f'(analytic)] with
    #different values of dx
    
plt.plot(np.log10(dxarr), np.log10(yarr))
plt.title("Forward Difference")
plt.xlabel("log10(dx)")
plt.ylabel("log10(error)")
plt.show()
```

#### Error

The error is approximately $dx^{-8}$. This means that n = -8. Also the optimal value for dx would be $10^{-8}$. This is because as the values grow larger than $10^{-8}$, the error will increase. This is because we can only get about half of the numerical precision of a derivative using forward difference. 

#### 2) repeat the exercise using the central difference scheme. Why this scheme is better

```python
def dericd(y,x,dx):
    return (y(x+(dx/2))-y(x-(dx/2)))/dx # page 191 in mark newman computational physics

x = np.pi/4
dxarr = []
yarr = []
for dx in 10**np.arange(-16, -1, 0.1):
    y = np.abs(dericd(np.cos,x,dx)+np.sin(x))
    dxarr.append(dx)
    yarr.append(y)
plt.plot(np.log10(dxarr), np.log10(yarr))
plt.title("Central Difference")
plt.xlabel("log10(dx)")
plt.ylabel("log10(error)")
plt.show()
```

#### Error

The error is approximately $dx^{-8}$. This means that the n = -8 and the optimal value for dx would be -5. With the central difference, the ideal value is around $C^{1/3}$ and this matches what is shown in the graph. This result is better because it has a higher order scheme, which means it gets a more accurate result with a bigger interval. The Central difference is more accurate by a factor of about 100 because the error would be around $C^{2/3}$ which shows in this data if we let C = -16.

#### 3) repeat (2) for x = 0. Why do the results differ?

```python
def dericd(y,x,dx):
    return (y(x+(dx/2))-y(x-(dx/2)))/dx # page 191 in mark newman computational physics

x = 0
dxarr = []
yarr = []
for dx in 10**np.arange(-16, -1, 0.1):
    y = np.abs(dericd(np.cos,x,dx)+np.sin(x))
    dxarr.append(dx)
    yarr.append(y)
plt.plot(dxarr,yarr)
plt.title("Central Difference")
plt.xlabel("dx")
plt.ylabel("error")
plt.show()
```

##### Why the results differ?

The result differs because what the central difference does is place two points symmetrically around x. one at $1/2$ h and the other at $-1/2h$. Since the value of x is zero, they are placed at $h/2, -h/2$. But since x is equal to zero, h = 0 as well since there would be no error.
