### Write a program to plot
$$f(x) = \pi/2 + 2 * (\sin(x)+\sin(3x)/3+sin(5x/5)+...sin(nx)/n$$
from $-\pi$ to $\pi$. Choose n = 11, 21, and 31 and plot all three curves in one figure.

#### The function:
This cell goes over the function sin(nx)/n. I assigned the variable y1 the value sin(x), then i did a for loop for when i was in range of the value n. Then i included an if statement for when the values are odd and i added sin(ix)/i to the value of y1. This made it easier to update the value of y1 according to the counter i. It also ignores 0 and avoids a division by 0 error. I created a const pi/2 since it had nothing to do with this iteration.

```python
const = pi/2
def f(x,n):
    y1 = sin(x)
    for i in range(n):
        if(i % 2 != 0):
            y1 += sin(i*x)/i
    return y1
```

#### The graphing
In order to graph the function i created, I first needed to store the values of y1 from the function to another arbitrary value so that i can plug it into plt.plot(). This was repeated for n=11,n=21,n=31. I then combined all the graphs into 1.

```python
y2 = const+2*f(x,11)
plt.plot(x,y2)
plt.title("n = 11")
plt.show()
```

```python
y3 = const + 2*f(x,21)
plt.plot(x,y3)
plt.title("n = 21")
plt.show()
```

```python
y4 = const+ 2*f(x,31)
plt.plot(x,y4)
plt.title("n = 31")
plt.show()
```
